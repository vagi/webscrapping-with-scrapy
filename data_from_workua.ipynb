```python
# Correlation between age and demanded salary in Kharkov (UA)
data.plot(kind='scatter', x='age', y='salary')
plt.show
```

```python
# Highest salaries required by IT specialists
df_highest_salary = data.sort_values(by=['salary'], ascending=False).head(10)   # na_position='first'
df_highest_salary.plot(kind='bar', x='position', y='salary')
plt.xticks(rotation=90)
```

```python
# Lowest salaries required by other specialists
df_lowest_salary = data.sort_values(by=['salary'], ascending=True).head(10)   # na_position='first'
df_lowest_salary.plot(kind='bar', x='position', y='salary')
plt.xticks(rotation=90)
```
